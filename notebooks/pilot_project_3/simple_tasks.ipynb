```python
# 4.4.2
def find_timedelta_in_days(a, b):
    temp_list = list()
    temp_list.append(a)
    temp_list.append(b)
    
    sorted_list = sorted(temp_list, reverse=True)
    first_date = sorted_list[0]
    second_date = sorted_list[1]
    diff_date = first_date - second_date
    # Total number of seconds between dates
    duration_in_s = diff_date.total_seconds()
    
    days = diff_date.days
    # Seconds in a day = 86400
    days = divmod(duration_in_s, 86400)[0]
    return days
```

```python
diff_list = list()
for index, row in df.iterrows():
    review_list = row['Reviews'].replace('[', '').replace(']', '').replace('\'', '').split(', ')
    
    r = re.compile("[01-9]{2}/[01-9]{2}/[01-9]{4}")
    review_dates = list(filter(r.match, review_list)) 
    
    dates_list = [datetime.strptime(date, '%m/%d/%Y').date() for date in review_dates]
    if len(dates_list) == 2:
        result = find_timedelta_in_days(dates_list[0], dates_list[1])
        diff_list.append(result)

max(diff_list)
```
